```python
from keras.datasets import mnist
from sklearn.cluster import MiniBatchKMeans
import numpy as np

from classy.kmeans import retrieve_info, print_metrics, get_predicted_labels

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train = x_train / 255.0
x_test = x_test / 255.0
X_train = x_train.reshape(len(x_train), -1)
X_test = x_test.reshape(len(x_test), -1)

total_clusters = len(np.unique(y_test))


kmeans = MiniBatchKMeans(
    n_clusters = 384,
    n_init = 8,
    batch_size = 1000
)
kmeans.fit(X_train)
predicted_labels = get_predicted_labels(kmeans, y_train)
print_metrics(kmeans, y_train, predicted_labels)
```

```python
from sklearn import metrics

reference_labels = retrieve_info(kmeans.labels_, y_train)

train_pred = np.array([reference_labels[c] for c in kmeans.labels_])
print("Train accuracy:", metrics.accuracy_score(y_train, train_pred))

test_clusters = kmeans.predict(X_test)
test_pred = np.array([reference_labels[c] for c in test_clusters])
print("Test accuracy:", metrics.accuracy_score(y_test, test_pred))
```

```python
from sklearn.cluster import KMeans

kmeans_full: KMeans = KMeans(n_clusters=384, n_init=8)
kmeans_full.fit(X_train)

predicted_labels = get_predicted_labels(kmeans_full, y_train)
print_metrics(kmeans_full, y_train, predicted_labels)
```

```python
train_pred = np.array([reference_labels[c] for c in kmeans_full.labels_])
print("Train accuracy:", metrics.accuracy_score(y_train, train_pred))

test_clusters = kmeans_full.predict(X_test)
test_pred = np.array([reference_labels[c] for c in test_clusters])
print("Test accuracy:", metrics.accuracy_score(y_test, test_pred))
```
